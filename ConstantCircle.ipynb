```python
swap10 = ConstantCircle_swap([Token('x', 100), Token('y', 100)], 10, 0)
swap20 = ConstantCircle_swap([Token('x', 100), Token('y', 100)], 20, 0)
swap30 = ConstantCircle_swap([Token('x', 100), Token('y', 100)], 30, 0)
swap350 = ConstantCircle_swap([Token('x', 100), Token('y', 100)], 350, 0)
swap500 = ConstantCircle_swap([Token('x', 100), Token('y', 100)], 500, 0)
swap700 = ConstantCircle_swap([Token('x', 100), Token('y', 100)], 700, 0)

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))
#x = [i in range(0, swap01.cost_function(), 1)]
x = np.arange(0, swap10.unknownAmount(0, swap10.getTokens()[1]), 1)
axes[0][0].plot(x, [swap10.unknownAmount(j, swap10.getTokens()[1]) for j in x], color = 'red', label = 'a')

x = np.arange(0, swap20.unknownAmount(0, swap20.getTokens()[1]), 1)
axes[0][1].plot(x, [swap20.unknownAmount(j, swap20.getTokens()[1]) for j in x], color = 'red', label = 'b')

x = np.arange(0, swap30.unknownAmount(0, swap30.getTokens()[1]),1)
axes[1][0].plot(x, [swap30.unknownAmount(j, swap30.getTokens()[1]) for j in x], color = 'red', label = 'c')

x = np.arange(0, swap350.unknownAmount(0, swap350.getTokens()[1]), 1)
axes[1][1].plot(x, [swap350.unknownAmount(j, swap350.getTokens()[1]) for j in x], color = 'red', label = 'd')

x = np.arange(0, swap500.unknownAmount(0, swap500.getTokens()[1]), 1)
axes[2][0].plot(x, [swap500.unknownAmount(j, swap500.getTokens()[1]) for j in x], color = 'red', label = 'd')

x = np.arange(0, swap700.unknownAmount(0, swap700.getTokens()[1]), 1)
axes[2][1].plot(x, [swap700.unknownAmount(j, swap700.getTokens()[1]) for j in x], color = 'red', label = 'd')

plt.savefig('ConstantCircle.jpeg')
plt.show()
```

```python
a = [10, 20, 30, 350, 500, 700]
```

```python
#Uticaj parametra a na kupovinu svih tokena

for a_i in a:
    cc = ConstantCircle_swap([Token('x', 100), Token('y', 100)], a_i, 0)
    c = round(cc.cost_function(), 4)
    p0_x = round(cc.price_function(cc.getTokens()[0]), 4)
    p0_y = round(cc.price_function(cc.getTokens()[1]), 4)
    pay = round(cc.swap('buy','x', 100), 4)
    p1_x = round(cc.price_function(cc.getTokens()[0]), 4)
    p1_y = round(cc.price_function(cc.getTokens()[1]), 4)
    
    print('a=', a_i, 'C=', c, 'P0(x)= ', p0_x, 'P0(y)= ', p0_y, 'pay=', pay,  'P1(x)= ', p1_x, 'P1(y)= ', p1_y)
    cc.poolState()
    print()
```

```python
#Kupovina jedan po jedan
swap_path_idepedent = ConstantCircle_swap([Token('x', 100), Token('y', 100)], 350, 0)
s = 0
for i in range(10):
    pay = round(swap_path_idepedent.swap('buy', 'x', 1), 5)
    print()
    print('Korisnik placa: ', round(pay, 4))
    swap_path_idepedent.poolState()
    s += pay
print()
print('Ukupni troskovi', round(s, 4))
```

Osetljivost na likvidnost

```python
init = [50, 80, 100, 1000, 5000, 10000]
```

```python
for i in init:
    swap_liquidity = ConstantCircle_swap([Token('x', i), Token('y', i)], 350, 0)
    print(i)
    print('c0 = ' ,round(swap_liquidity.cost_function(), 4))
    print('p1_0 = ', round(swap_liquidity.price_function(swap_liquidity.getTokens()[0]), 4))
    print('p2_0 = ', round(swap_liquidity.price_function(swap_liquidity.getTokens()[1]), 4))
    pay = swap_liquidity.swap('buy','x', 10)
    print('pay = ', round(pay, 4))
    print('p1_1 = ', round(swap_liquidity.price_function(swap_liquidity.getTokens()[0]), 4))
    print('p2_1 = ', round(swap_liquidity.price_function(swap_liquidity.getTokens()[1]), 4))
    
    swap_liquidity.poolState()
    print()
```

```python
swap50 = ConstantCircle_swap([Token('x', 50), Token('y', 50)], 350, 0)
swap80 = ConstantCircle_swap([Token('x', 80), Token('y', 80)], 350, 0)
swap100 = ConstantCircle_swap([Token('x', 100), Token('y', 100)], 350, 0)
swap1000 = ConstantCircle_swap([Token('x', 1000), Token('y', 1000)], 350, 0)
swap5000 = ConstantCircle_swap([Token('x', 5000), Token('y', 5000)], 350, 0)
swap10000 = ConstantCircle_swap([Token('x', 10000), Token('y', 10000)], 350, 0)

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))
#x = [i in range(0, swap01.cost_function(), 1)]
x = np.arange(0, swap50.unknownAmount(0, swap50.getTokens()[1]), 1)
axes[0][0].plot(x, [swap50.unknownAmount(j, swap50.getTokens()[1]) for j in x], color = 'green', label = 'a')

x = np.arange(0, swap80.unknownAmount(0, swap80.getTokens()[1]), 1)
axes[0][1].plot(x, [swap80.unknownAmount(j, swap80.getTokens()[1]) for j in x], color = 'green', label = 'b')

x = np.arange(0, swap100.unknownAmount(0, swap100.getTokens()[1]),1)
axes[1][0].plot(x, [swap100.unknownAmount(j, swap100.getTokens()[1]) for j in x], color = 'green', label = 'c')

x = np.arange(0, swap1000.unknownAmount(0, swap1000.getTokens()[1]), 1)
axes[1][1].plot(x, [swap1000.unknownAmount(j, swap1000.getTokens()[1]) for j in x], color = 'green', label = 'd')

x = np.arange(0, swap5000.unknownAmount(0, swap5000.getTokens()[1]), 1)
axes[2][0].plot(x, [swap5000.unknownAmount(j, swap5000.getTokens()[1]) for j in x], color = 'green', label = 'd')

x = np.arange(0, swap10000.unknownAmount(0, swap10000.getTokens()[1]), 1)
axes[2][1].plot(x, [swap10000.unknownAmount(j, swap10000.getTokens()[1]) for j in x], color = 'green', label = 'd')

plt.savefig('ConstantCircleLiquidity.jpeg')
plt.show()
```
